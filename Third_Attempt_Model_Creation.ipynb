```python
# Drop the non-beneficial ID columns, just 'EIN' for now.
newapp3_df=application3_df.drop(['EIN'], axis=1)
newapp3_df.head()
```

```python
# Look at APPLICATION_TYPE value counts again for binning
newapp3_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# application_types_to_replace=[0,156,528,725,737,1065,1173, 1216,1542, 27037]        

# # Replace in dataframe
# for app in application_types_to_replace:
#     application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# # Check to make sure binning was successful
# application_df['APPLICATION_TYPE'].value_counts()

#Code obtained from RG who got it from Bill
newapp3_df = newapp3_df.apply(lambda x: x.mask(x.map(x.value_counts())<10, 'Other') if x.name=='APPLICATION_TYPE' else x)
newapp3_df['APPLICATION_TYPE'].value_counts()
```

```python
#Check value counts for different instances of Classification again before we bin
list_classification3=newapp3_df['CLASSIFICATION'].value_counts()
list_class3=list_classification3.tolist()
list_class3
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# # use the variable name `classifications_to_replace`
# #  YOUR CODE GOES HERE

# # Replace in dataframe
# for cls in classifications_to_replace:
#     application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# # Check to make sure binning was successful
# application_df['CLASSIFICATION'].value_counts()
newapp_df3 = newapp3_df.apply(lambda x: x.mask(x.map(x.value_counts())<100, 'Other') if x.name=='CLASSIFICATION' else x)
newapp_df3['CLASSIFICATION'].value_counts()
```

```python
#check value counts again for 'name' column to then preview how we should bin
newapp3_df['NAME'].value_counts().tolist()
```

```python
newapp_df3 = newapp3_df.apply(lambda x: x.mask(x.map(x.value_counts())<140, 'Other') if x.name=='NAME' else x)
newapp_df3['NAME'].value_counts()
```

```python
# Convert categorical data to numerical data with `pd.get_dummies`
dummy_df3=pd.get_dummies(newapp_df3)
dummy_df3
```

```python
# Drop Special_Considerations_N column because it tells us the same thing as Special_Considerations_Y
#e.g. value of 1 in Special_Considerations_Y is the same as value of 0 in Special_Considerations_N
newdummy_df3=dummy_df3.drop(['SPECIAL_CONSIDERATIONS_N'], axis=1)
newdummy_df3
```

```python
# Split our preprocessed data into our features and target arrays

#Target variable is going to be whether or not the applicants applying for funding will be successful, as we can see in 
#the possible binary result of "1" or "0" in the "IS SUCCESSFUL" column. This is what we will be pulling out to to target
#for the model that we are going to create. All other columns will be our features which will be used to predict the 
#'IS_SUCCESSFUL result'.

y=newdummy_df3['IS_SUCCESSFUL'].values
X=newdummy_df3.drop(columns='IS_SUCCESSFUL').values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn3_model = tf.keras.models.Sequential()

# First hidden layer
nn3_model.add(tf.keras.layers.Dense(units=25, activation="relu", input_dim=133))

#Second hidden layer
nn3_model.add(tf.keras.layers.Dense(units=20, activation="relu"))

# #Third hidden layer
nn3_model.add(tf.keras.layers.Dense(units=15, activation="relu"))


# Output layer
nn3_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn3_model.summary()
```

```python
# Compile the model
nn3_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn3_model.fit(X_train_scaled, y_train, epochs=25)
#three lines below SHOW results from running different epochs
# Loss: 0.506847083568573, Accuracy: 0.7516034841537476 WINNER WINNER!
#50 epochs = Loss: 0.5117678046226501, Accuracy: 0.7496209740638733
#75 epochs = Loss: 0.5185121893882751, Accuracy: 0.7486880421638489
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn3_model.evaluate(X_test_scaled,y_test,verbose=4)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
