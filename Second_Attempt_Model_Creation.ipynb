## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
import pickle

#  Import and read the charity_data.csv.
import pandas as pd 
application2_df = pd.read_csv("charity_data.csv")
application2_df
```

```python
application2_df['NAME'].value_counts()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
newapp2_df=application2_df.drop(['EIN','NAME'], axis=1)
newapp2_df.head()

#Target variable is going to be whether or not the applicants applying for funding will be successful, as we can see in 
#the possible binary result of "1" or "0" in the "IS SUCCESSFUL" column. This is what we will be pulling out to to target
#for the model that we are going to create
```

```python
#lets look for important features
from sklearn.feature_selection import SelectFromModel
```

```python
# Look at APPLICATION_TYPE value counts for binning
newapp2_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# application_types_to_replace=[0,156,528,725,737,1065,1173, 1216,1542, 27037]        

# # Replace in dataframe
# for app in application_types_to_replace:
#     application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# # Check to make sure binning was successful
# application_df['APPLICATION_TYPE'].value_counts()

#Code obtained from RG who got it from Bill
newapp2_df = newapp2_df.apply(lambda x: x.mask(x.map(x.value_counts())<10, 'Other') if x.name=='APPLICATION_TYPE' else x)
newapp2_df['APPLICATION_TYPE'].value_counts()
```

```python
#Check value counts for different instances of Classification
list_classification2=newapp2_df['CLASSIFICATION'].value_counts()
list_class2=list_classification2.tolist()
list_class2
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# # use the variable name `classifications_to_replace`
# #  YOUR CODE GOES HERE

# # Replace in dataframe
# for cls in classifications_to_replace:
#     application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# # Check to make sure binning was successful
# application_df['CLASSIFICATION'].value_counts()
newapp_df2 = newapp2_df.apply(lambda x: x.mask(x.map(x.value_counts())<100, 'Other') if x.name=='CLASSIFICATION' else x)
newapp_df2['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummy_df2=pd.get_dummies(newapp_df2)
dummy_df2.columns
```

```python
# Drop Special_Considerations_N column because it tells us the same thing as Special_Considerations_Y
# A value of 1 in Special_Considerations_Y column means YES
newdummy_df2=dummy_df2.drop(['SPECIAL_CONSIDERATIONS_N'], axis=1)
newdummy_df2
```

```python
# Split our preprocessed data into our features and target arrays
y=newdummy_df2['IS_SUCCESSFUL'].values
X=newdummy_df2.drop(columns='IS_SUCCESSFUL').values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn2_model = tf.keras.models.Sequential()

# First hidden layer
nn2_model.add(tf.keras.layers.Dense(units=100, activation="relu", input_dim=52))

#Second hidden layer
nn2_model.add(tf.keras.layers.Dense(units=15, activation="relu"))

#Third hidden layer
nn2_model.add(tf.keras.layers.Dense(units=10, activation="relu"))


# Output layer
nn2_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn2_model.summary()
```

```python
# Compile the model
nn2_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn2_model.fit(X_train_scaled, y_train, epochs=25)
#23 epochs yielded .726
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn2_model.evaluate(X_test_scaled,y_test,verbose=4)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
